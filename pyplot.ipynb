1. Given the text below, create a bar plot (or a horizontal bar plot) that shows the frequency distribution of characters in the string. 
A dictionary has been created for you containing the characters and the frequencies of the characters.
Make sure your bar plot has the bars labelled correctly (with the character it represents the frequency of).

        Answer: I will use Mathplotlib to plot this data. I will define  the x-axis and y-axis values as lists. Plot the bar, give a name to x-axis and y-axis using .xlabel() and .ylabel() functions, give title to the plot using .title() function, and finally, to view the plot, I will use .show() function.

2. Generate a sequence of at least 5 random numbers (integers are fine), and visualize the data using both `scatter` and `plot` on different axes.  You may use the index of the value as its x value in the plot.

        Answer: I will create a scatter plot in Mathplotlib, using the "scatter" function provided by the library.  This function takes two arrays of data points- one for the x-axis and one for the y-axis- and plots them as invididual points on the graph.  Then I will create a line plot using the plt.plot() method where the first argument is x variable and the second argument is the y- variable

```python
import numpy as np
import matplotlib.pyplot as plt

#Generating 5 randome intigers between 0 and 50 using random from numpy.
random_numbers=np.random.randint(1, 50, size=5)

#Creating the scatter plot
fig1=plt.scatter(range(len(random_numbers)), random_numbers)
plt.title('Random Number Scatter Graph')
plt.figure(figsize=(9, 4))
print(fig1)

#Creating the graph using plot
fig2=plt.plot(range(len(random_numbers)), random_numbers)
plt.title('Random Number Plot')
plt.figure(figsize=(9, 4))
print(fig2)
```

3. Do the same question as above, but change the syle and color of both the scatter and the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

#Generating 5 randome intigers between 1 and 50 using random from numpy.
random_numbers=np.random.randint(1, 50, size=5)

#Creating the scatter plot with different shape
fig3=plt.scatter(range(len(random_numbers)), random_numbers, marker='o', color='green')
plt.title('Scatter with Random Number Numpy')
plt.figure(figsize=(9, 4))
print(fig3)

#Creating a plot using a different line style
fig4=plt.plot(range(len(random_numbers)), random_numbers, linestyle='--', linewidth=0.5, color='red')
plt.title('Plot with Random Number Numpy')
plt.figure(figsize=(9, 4))
print(fig4)
```

4. Given the data sets below, plot or scatter both on the same set of axes.  You are plotting the time it takes to execute a sorting algorithm on a list of a given size; your x axis should be the size, and the y axis is time.  Assume that `insertion_sort_times[i]` is how long it took to sort a list of length `sizes[i]`.  Add a legend to your plot showing which is insertion sort and which is merge sort.  Use the variable names to determine which data is which. 

        Answer: I used similar approach to that in Questions 1 and 2

```python
# data given:  
# sizes = [10, 20, 50, 100, 1000, 5000]
# insertion_sort_times = [0.0010252999999999998, 0.0027026999999999954, 0.010147200000000002, 0.0381137, 3.6303399, 91.2180796]
# merge_sort_times = [0.00161889999999687, 0.003635600000009731, 0.0090655000000055, 0.020108000000007564, 0.2687474000000094, 1.6147050999999948]

import matplotlib.pyplot as plt 

sizes = [10, 20, 50, 100, 1000, 5000] 

insertion_sort_times = [0.0010252999999999998, 0.0027026999999999954, 0.010147200000000002, 0.0381137, 3.6303399, 91.2180796] 

merge_sort_times = [0.00161889999999687, 0.003635600000009731, 0.0090655000000055, 0.020108000000007564, 0.2687474000000094, 1.6147050999999948] 

plt.figure(figsize=(9, 4))

plt.plot(sizes, insertion_sort_times, marker='o', linestyle='--', color='red', label='Insertion Sort Times') 

plt.plot(sizes, merge_sort_times, marker='x', linestyle='solid', color='green', label='Merge Sort Times') 

plt.xlabel('Size') 
plt.ylabel('Time (Seconds)') 
plt.title('Sorting Algorithm')
plt.legend() 

plt.show() 

```
